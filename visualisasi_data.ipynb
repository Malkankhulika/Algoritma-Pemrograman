```python
# Tugas Praktikum
# kHULIKA MALKAN (2311110057)
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des']
indonesia = [19,27,34,34,69,96,117,134,172,227,311,369]
japan = [7,11,15,20,94,105,122,147,159,170,189,214]

x = np.arange(len(bulan))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 7))

indo = ax.bar(x - width/2, indonesia, width, label='Indonesia', color='orange')
japan = ax.bar(x + width/2, japan, width, label='Japan', color='green')

ax.set_title('Penyebaran Kasus Covid di Indonesia dan Japan', size=16)
ax.set_ylabel('Jumlah Kasus', size=14)
ax.set_xticks(x)
ax.set_xticklabels(bulan, size=12)
ax.legend(fontsize=14)

plt.show()
```
