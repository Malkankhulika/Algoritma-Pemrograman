```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat nilai x dari -5 hingga 5
x = np.linspace(-5, 5, 400)

# Menghitung nilai y menggunakan persamaan parabola y^2 = 16x
y1 = np.sqrt(16 * x)
y2 = -np.sqrt(16 * x)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = √(16x)')
plt.plot(x, y2, label='y = -√(16x)')
plt.title('Grafik Parabola y^2 = 16x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat nilai y dari -5 hingga 0
y = np.linspace(-5, 0, 400)

# Menghitung nilai x menggunakan persamaan irisan kerucut x^2 = -16y
x = np.sqrt(-16 * y)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='x^2 = -16y')
plt.plot(-x, y, label='-x^2 = -16y')
plt.title('Grafik Irisan Kerucut berbentuk Parabola')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat nilai x dari -10 hingga 10
x = np.linspace(-10, 10, 400)

# Menghitung nilai y menggunakan persamaan hiperbola x^2/10 - y^2/4 = 1
y1 = np.sqrt((x**2 / 10 - 1) * 4)
y2 = -np.sqrt((x**2 / 10 - 1) * 4)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = √((x^2/10) - 1)')
plt.plot(x, y2, label='y = -√((x^2/10) - 1)')
plt.plot(x, -y1, color='blue')  # plot y = -√((x^2/10) - 1) di bagian bawah juga
plt.plot(x, -y2, color='blue')  # plot y = √((x^2/10) - 1) di bagian bawah juga
plt.title('Grafik Hiperbola x^2/10 - y^2/4 = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat nilai sudut theta dari 0 hingga 2*pi
theta = np.linspace(0, 2*np.pi, 100)

# Persamaan parametrik untuk elips
a = 3  # akar dari angka yang ada di bawah x^2
b = 4  # akar dari angka yang ada di bawah y^2
x = a * np.cos(theta)
y = b * np.sin(theta)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Irisan Kerucut')
plt.title('Grafik Irisan Kerucut berbentuk Elips')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.axis('equal')
plt.show()
```
