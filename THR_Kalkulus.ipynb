```python
# Nama : kHULIKA MALKAN
# NIM  : 2311110057


import numpy as np
import matplotlib.pyplot as plt

# Persamaan parabola
def parabola(y):
    return 1/16 * y**2

# Membuat array titik y
y = np.linspace(-10, 10, 400)

# Menghitung nilai x berdasarkan fungsi parabola
x = parabola(y)

# Plot parabola
plt.plot(x, y, label='Parabola: $y^2 = 16x$')

# Menambahkan fokus
plt.plot(4, 0, 'ro', label='Fokus (4, 0)')

# Menambahkan garis direktris
plt.axvline(x=-4, color='g', linestyle='--', label='Garis Direktris x = -4')

# Menambahkan titik latus rectum
plt.plot(4, 8, 'bo', label='Titik Latus Rectum (4, 8)')
plt.plot(4, -8, 'bo', label='Titik Latus Rectum (4, -8)')

# Menambahkan sumbu simetri
plt.axhline(y=0, color='m', linestyle='-.', label='Sumbu Simetri y = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('GRAFIK PARABOLA  $y^2 = 16x$')
plt.legend(fontsize = 'xx-small', loc = 'lower left')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar elips x^2/9 + y^2/16 = 1
def plot_ellipse():
    # Buat array theta dari 0 hingga 2 * pi (sudut penuh elips)
    theta = np.linspace(0, 2 * np.pi, 100)

    # Tentukan nilai a (sumbu semi-mayor) dan b (sumbu semi-minor)
    a = 4  # akar dari 16
    b = 3  # akar dari 9

    # Hitung koordinat x dan y untuk elips
    x = a * np.cos(theta)
    y = b * np.sin(theta)

    # Buat plot
    plt.figure()

    # Plot elips
    plt.plot(x, y, label='x^2/9 + y^2/16 = 1')

    # Tambahkan label sumbu dan judul
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Kurva x^2/9 + y^2/16 = 1 (Elips)')
    plt.legend()

    # Tampilkan plot
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')  # Untuk menjaga proporsi elips
    plt.show()

# Panggil fungsi untuk menggambar elips x^2/9 + y^2/16 = 1
plot_ellipse()
```
